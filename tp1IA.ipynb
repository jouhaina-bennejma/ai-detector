Parcours en **largeur**

```python
import queue  # Importation du module queue pour utiliser la file d'attente

def bfs(graph, start):
    # Initialisation de la file d'attente et de l'ensemble des nœuds visités
    q = queue.Queue()  # Création de la file d'attente
    visited = set()  # Ensemble pour suivre les nœuds déjà visités
    visited_order = []  # Liste pour stocker l'ordre de visite des nœuds

    # Ajouter le nœud de départ dans la file et le marquer comme visité
    q.put(start)
    visited.add(start)

    while not q.empty():  # Tant que la file n'est pas vide
        # Défile le premier nœud de la file
        node = q.get()
        visited_order.append(node)  # Ajouter ce nœud à la liste des visites

        # Explorer tous les voisins du nœud actuel
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)  # Marquer le voisin comme visité
                q.put(neighbor)  # Ajouter le voisin dans la file d'attente pour exploration

    return visited_order

# Exemple d'utilisation avec un graphe représenté sous forme de dictionnaire
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

start = 'A'

# Appeler la fonction bfs pour obtenir l'ordre de visite des nœuds
visited_order = bfs(graph, start)

# Afficher l'ordre des nœuds visités lors du parcours en largeur
print("Ordre de visite des nœuds (parcours en largeur) :", ' -> '.join(visited_order))
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def parcours_largeur(G, s):
    """
    Effectue un parcours en largeur (BFS) dans un graphe à partir du sommet s,
    sans utiliser de liste 'visites' séparée.

    Arguments:
    - G : un dictionnaire représentant le graphe. Chaque clé est un noeud et
          la valeur est une liste de ses voisins.
    - s : le noeud à partir duquel commencer le parcours.

    Retourne une liste des noeuds visités dans l'ordre du parcours.
    """
    # Liste qui servira de "file" pour gérer les noeuds à explorer
    file = [s]

    # Liste pour stocker l'ordre des noeuds visités
    ordre_visites = []

    while file:
        # Défiler un noeud de la file (en enlevant le premier élément)
        s = file.pop(0)

        # Si le noeud n'a pas encore été visité, on le traite
        if s not in ordre_visites:
            # Traiter le noeud actuel (ici, on l'ajoute à la liste des visités)
            ordre_visites.append(s)

            # Explorer tous les voisins du noeud s
            for t in G[s]:
                # Ajouter les voisins à la file si ce n'est pas déjà fait
                if t not in ordre_visites:
                    file.append(t)

    return ordre_visites

# Exemple d'utilisation du code avec un graphe simple
graphe = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Appel du parcours en largeur à partir du noeud 'A'
resultat_bfs = parcours_largeur(graphe, 'A')
print("Ordre des noeuds visités (parcours en largeur):", resultat_bfs)

# Affichage du graphe avec matplotlib et networkx
G = nx.Graph(graphe)  # Créer un graphe à partir du dictionnaire

# Dessiner le graphe
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='skyblue', font_size=10, node_size=2000, font_weight='bold', edge_color='gray')
plt.title("Affichage du Graphe")
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def parcours_profondeur(G, s):
    """
    Effectue un parcours en profondeur (DFS) dans un graphe à partir du sommet s.

    Arguments:
    - G : un dictionnaire représentant le graphe. Chaque clé est un noeud et
          la valeur est une liste de ses voisins.
    - s : le noeud à partir duquel commencer le parcours.

    Retourne une liste des noeuds visités dans l'ordre du parcours.
    """
    # Liste qui servira de "pile" pour gérer les noeuds à explorer
    pile = [s]

    # Liste pour stocker l'ordre des noeuds visités
    ordre_visites = []

    while pile:
        # Dépiler un noeud de la pile (en enlevant le dernier élément)
        s = pile.pop()

        # Si le noeud n'est pas encore visité, on le traite
        if s not in ordre_visites:
            # Traiter le noeud actuel (ici, on l'ajoute à la liste des visités)
            ordre_visites.append(s)

            # Explorer tous les voisins du noeud s, on les empile
            for t in G[s]:
                # Empiler le voisin t dans la liste (ajouter à la fin de la "pile")
                if t not in ordre_visites:
                    pile.append(t)

    return ordre_visites

# Exemple d'utilisation du code avec un graphe simple
graphe = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Appel du parcours en profondeur à partir du noeud 'A'
resultat_dfs = parcours_profondeur(graphe, 'A')
print("Ordre des noeuds visités (parcours en profondeur):", resultat_dfs)

# Affichage du graphe avec matplotlib et networkx
G = nx.Graph(graphe)  # Créer un graphe à partir du dictionnaire

# Dessiner le graphe
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='skyblue', font_size=10, node_size=2000, font_weight='bold', edge_color='gray')
plt.title("Affichage du Graphe")
plt.show()
```

```python
import queue

def parcours_profondeur(graphe, start):
    """
    Effectue un parcours en profondeur dans un graphe en utilisant une pile.

    Arguments:
    - graphe : un dictionnaire représentant le graphe. Chaque clé est un noeud et
               la valeur est une liste de ses voisins.
    - start : le noeud à partir duquel commencer le parcours.

    Retourne une liste des noeuds visités dans l'ordre du parcours.
    """
    # Instanciation d'une pile vide
    p = queue.LifoQueue()

    # Initialisation d'un ensemble pour garder trace des noeuds visités
    visites = set()

    # Mettre le noeud de départ dans la pile
    p.put(start)

    # Liste pour stocker l'ordre des noeuds visités
    ordre_visites = []

    while not p.empty():
        # Dépiler un noeud de la pile
        x = p.get()

        # Si ce noeud n'a pas encore été visité
        if x not in visites:
            # Marquer comme visité
            visites.add(x)

            # Ajouter ce noeud à la liste des noeuds visités
            ordre_visites.append(x)

            # Ajouter tous les voisins non visités du noeud actuel dans la pile
            for voisin in graphe[x]:
                if voisin not in visites:
                    p.put(voisin)

    return ordre_visites

# Exemple d'utilisation du code avec un graphe simple
graphe = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Appel du parcours en profondeur à partir du noeud 'A'
resultat = parcours_profondeur(graphe, 'A')
print("Ordre des noeuds visités:", resultat)
```
