```python
import networkx as nx

# Cr√©ation du graphe
G = nx.Graph()
G.add_weighted_edges_from([
    ('A', 'B', 1),
    ('A', 'C', 4),
    ('B', 'C', 2),
    ('B', 'D', 5),
    ('C', 'D', 1)
])


# Demander √† l'utilisateur les points de d√©part et d'arriv√©e
start = input("Donne le point de d√©part : ").strip().upper()
end = input("Donne le point d'arriv√©e : ").strip().upper()

# V√©rifier si les n≈ìuds existent dans le graphe
if start not in G.nodes or end not in G.nodes:
    print("Erreur: Un des points n'existe pas dans le graphe.")
else:
    # Trouver le chemin et le co√ªt minimal
    chemin = nx.shortest_path(G, source=start, target=end, weight='weight')
    cout_min = nx.shortest_path_length(G, source=start, target=end, weight='weight')

    print("‚û°Ô∏è Chemin le plus court:", " ‚Üí ".join(chemin))
    print("üí∞ Co√ªt minimal:", cout_min)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Cr√©ation du graphe avec des poids
G = nx.Graph()
edges = [
    ('A', 'B', 1), ('A', 'C', 4),
    ('B', 'C', 2), ('B', 'D', 5),
    ('C', 'D', 1)
]
G.add_weighted_edges_from(edges)

# Demander √† l'utilisateur les points de d√©part et d'arriv√©e
start = input("Donne le point de d√©part : ").strip().upper()
end = input("Donne le point d'arriv√©e : ").strip().upper()

# V√©rifier si les n≈ìuds existent
if start not in G.nodes or end not in G.nodes:
    print("‚ùå Erreur: Un des points n'existe pas dans le graphe.")
else:
    # Trouver le chemin et le co√ªt minimal
    chemin = nx.shortest_path(G, source=start, target=end, weight='weight')
    cout_min = nx.shortest_path_length(G, source=start, target=end, weight='weight')

    print("\n‚û°Ô∏è Chemin le plus court:", " ‚Üí ".join(chemin))
    print("üí∞ Co√ªt minimal:", cout_min)

    # Dessiner le graphe
    pos = nx.spring_layout(G)  # Position des n≈ìuds
    plt.figure(figsize=(8,6))

    # Dessiner tous les ar√™tes
    nx.draw(G, pos, with_labels=True, node_size=2000, node_color="lightblue", font_size=12, font_weight="bold", edge_color="gray")

    # Dessiner les poids
    labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

    # Mettre en √©vidence le plus court chemin
    edges_path = list(zip(chemin, chemin[1:]))  # Transformer le chemin en liste d'ar√™tes
    nx.draw_networkx_edges(G, pos, edgelist=edges_path, edge_color="red", width=2)

    plt.title(f"Chemin le plus court de {start} √† {end}")
    plt.show()
```
